```python
import numpy as np
import matplotlib.pyplot as plt

from part1helpers import simulate_many

np.random.seed(0)
drift_rate_mean = -0.04  # v
drift_rate_variance = 0.04  # s
boundary_separation = 0.15  # a
starting_point = boundary_separation * 0.75  # z
time_step = 0.001  # dt
max_steps = 2000
```

```python
results = simulate_many(num_simulations=1,
                        time_step=time_step,
                        starting_point=boundary_separation*0.5,
                        drift_rate_mean=drift_rate_mean,
                        drift_rate_variance=drift_rate_variance,
                        boundary_separation=boundary_separation,
                        max_steps=max_steps
                        )

results_with_prior = simulate_many(num_simulations=1,
                        time_step=time_step,
                        starting_point=boundary_separation*0.75,
                        drift_rate_mean=drift_rate_mean,
                        drift_rate_variance=drift_rate_variance,
                        boundary_separation=boundary_separation,
                        max_steps=max_steps
                        )

for _, W_t in results:
    plt.plot(W_t, 'brown', label='Without prior')

for _, W_t in results_with_prior:
    plt.plot(W_t, 'green', label='With prior')

plt.xlabel("Timestep")
plt.ylabel("W_t")
plt.axhline(y=boundary_separation, color='rosybrown', linestyle='--')
plt.axhline(y=0, color='rosybrown', linestyle='--')
plt.legend()
plt.savefig('report/figures/part1/task3/example_paths_with_priors.pdf')
```

```python
results = simulate_many(num_simulations=1000,
                        time_step=time_step,
                        starting_point=starting_point,
                        drift_rate_mean=drift_rate_mean,
                        drift_rate_variance=drift_rate_variance,
                        boundary_separation=boundary_separation,
                        max_steps=max_steps
                        )
```

```python
correct_count = len([1 for hypothesis, W_t in results if hypothesis=="h_pos"])
incorrect_count = len([1 for hypothesis, W_t in results if hypothesis=="h_neg"])
timeout_count = len([1 for hypothesis, W_t in results if hypothesis==""])
print(correct_count/1000)
print(incorrect_count/1000)
print(timeout_count/1000)
```
