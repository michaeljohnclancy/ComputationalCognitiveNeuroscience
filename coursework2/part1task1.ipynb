```python
import numpy as np
import matplotlib.pyplot as plt

from part1helpers import simulate_many

np.random.seed(0)
drift_rate_mean = -0.04  # v
drift_rate_variance = 0.04  # s
boundary_separation = 0.15  # a
starting_point = boundary_separation / 2  # z
time_step = 0.001  # dt
max_steps = 2000
```

```python
results = simulate_many(num_simulations=10,
                        time_step=time_step,
                        starting_point=starting_point,
                        drift_rate_mean=drift_rate_mean,
                        drift_rate_variance=drift_rate_variance,
                        boundary_separation=boundary_separation,
                        max_steps=max_steps
                        )

for _, W_t in results:
    plt.plot(W_t)

plt.xlabel("Timestep")
plt.ylabel("W_t")
plt.axhline(y=boundary_separation, color='rosybrown', linestyle='--')
plt.axhline(y=0, color='rosybrown', linestyle='--')
plt.savefig('report/figures/part1/task1/example_paths.pdf')
```

```python
results = simulate_many(num_simulations=1000,
                        time_step=time_step,
                        starting_point=starting_point,
                        drift_rate_mean=drift_rate_mean,
                        drift_rate_variance=drift_rate_variance,
                        boundary_separation=boundary_separation,
                        max_steps=max_steps
                        )
```

```python
h_pos_times = []
for hypothesis, W_t in results:
    if hypothesis == "h_pos":
        h_pos_times.append(len(W_t))

plt.xlabel("Response Time")
plt.ylabel("Frequency")
plt.hist(h_pos_times, bins=20)
plt.savefig('report/figures/part1/task1/positive_response_time_histogram.pdf')
```

```python
h_neg_times = []
for hypothesis, W_t in results:
    if hypothesis == "h_neg":
        h_neg_times.append(len(W_t))

plt.xlabel("Response Time")
plt.ylabel("Frequency")
plt.hist(h_neg_times, bins=20)
plt.savefig('report/figures/part1/task1/negative_response_time_histogram.pdf')
```

```python
h_times = []
for hypothesis, W_t in results:
    h_times.append(len(W_t))

plt.xlabel("Response Time")
plt.ylabel("Frequency")
plt.hist(h_times, bins=20)
plt.savefig('report/figures/part1/task1/overall_response_time_histogram.pdf')
```
