```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from part2.part2helpers import simulate_trials

num_simulations = 100

# Model parameters
learning_rate = 0.35
inverse_temperature = 5.5
```

```python
all_stimulus_values = []
all_choices = []
all_reward_received = []
for i in range(num_simulations):
    stimulus_values, choices, reward_received = simulate_trials(learning_rate=learning_rate, inverse_temperature=inverse_temperature)
    all_stimulus_values.append(stimulus_values)
    all_choices.append(choices)
    all_reward_received.append(reward_received)

average_stimulus_values = pd.concat(all_stimulus_values).groupby(level=0).mean()
```

```python
plt.plot(np.arange(240), average_stimulus_values['A'], label='Stimulus A', c='brown')
plt.plot(np.arange(240), average_stimulus_values['B'], label='Stimulus B', c='green')
plt.xlabel("Trial number")
plt.ylabel("Average Stimulus Value")
plt.axvline(24, color='rosybrown', linestyle='--')
plt.axvline(48, color='rosybrown', linestyle='--')
plt.savefig('../report/figures/part2/task2/average_stimulus_values.pdf')
```

```python
average_stimulus_value_difference = (average_stimulus_values['A'] - average_stimulus_values['B']).abs()
plt.plot(np.arange(240), average_stimulus_value_difference,c='green')
plt.xlabel('Trial number')
plt.ylabel("Average Stimulus Value Difference")
plt.savefig('../report/figures/part2/task2/average_stimulus_value_difference.pdf')
```
